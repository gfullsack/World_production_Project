```python
import pandas as pd
from sqlalchemy import create_engine
import json
from pymongo import MongoClient
```

```python
client = MongoClient('localhost', 27017)
db = client['SF_Restaurant_db']
```

```python
csv_file = "Resources/FAO.csv"
Raw_Master_df = pd.read_csv(csv_file , encoding='latin-1')
Raw_Master_df.head(5)
```

```python
wine_df = Raw_Master_df[Raw_Master_df["Item"].str.match("Wine")]
wine_df
```

```python
Alcoholic_Beverages_df = Raw_Master_df[Raw_Master_df["Item"].str.match("Alcoholic Beverages")]
Alcoholic_Beverages_df
```

```python
Booze_df = pd.concat([Alcoholic_Beverages_df , wine_df], ignore_index = True)
```

```python
Master_Booze_df = Booze_df.sort_values('Area')
```
